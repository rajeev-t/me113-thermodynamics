```python
# dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# x-axis
out_temp = np.arange(0, 11, step=1)

# y-axis
q_dot = []
for temp in out_temp:
    q_dot.append(round(3840 - (153.6 * temp), 2))
```

```python
# create line graph expressing q_dot as a function of out_temp
plt.plot(out_temp, q_dot, c="darkred", marker=".")

# title/labels
plt.title("Rate of Heat Loss vs. Outside Temperature")
plt.xlabel("Outside Temperature (Â°C)")
plt.ylabel("Rate of Heat Loss (W)")

plt.xticks(out_temp)
plt.grid(True)

plt.savefig("resources/graph-1.png")

plt.show()
```
